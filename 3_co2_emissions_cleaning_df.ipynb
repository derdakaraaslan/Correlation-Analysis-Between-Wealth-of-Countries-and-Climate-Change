```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
path ='./CO2_emission.csv'
df_co2 = pd.read_csv(path)
```

```python
df = df_co2.drop(['1990', '1991', '1992', '1993', '1994', '2019.1'], axis=1)
df = pd.melt(df, id_vars=["Country Name", "country_code", "Region", "Indicator Name"],
        var_name="Year", 
        value_name="CO2").sort_values(by=['Country Name', 'Year'])
df['Year'] = df['Year'].astype(int)
```

Checking for null and duplicates values

```python
print("The missing values are:\n{}".format(df.isnull().sum()))
print("\nThe duplicated values are:{}".format(len(df[df.duplicated()])))
```

Removing null and duplicate values from dataset

```python
df.dropna(inplace = True)
print("The missing values are:\n{}".format(df.isnull().sum()))
print("\nThe duplicated values are:{}".format(len(df[df.duplicated()])))
```

Removing outliners

```python
for i in df[['CO2']]:
    plt.figure()
    sns.boxplot(df, x=df[i], orient="h")
```

```python
df = df[df['CO2'] < 10]
df
```

```python
df.to_csv('cleaned_CO2_emission.csv', index=False)
```
