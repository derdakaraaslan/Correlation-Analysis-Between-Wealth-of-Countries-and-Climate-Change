```python
import pandas as pd
import numpy as np 
```

```python
df = pd.read_csv('world_country_gdp_usd.csv') 
df.info()
```

```python
df.describe()
```

```python
df=df.rename(columns={"year": "Year"}).sort_values(by=['Country Name', 'Year'])
df
```

Checking for null and duplicates values

```python
print("The missing values are:\n{}".format(df.isnull().sum()))
print("\nThe duplicated values are:{}".format(len(df[df.duplicated()])))
```

```python
df.drop('Country Code',axis=1, inplace=True)
df.drop('GDP_USD',axis=1, inplace=True)
df.dropna(how="any", inplace=True)
print("The missing values are:\n{}".format(df.isnull().sum()))
print("\nThe duplicated values are:{}".format(len(df[df.duplicated()])))
```
